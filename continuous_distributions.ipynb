# Some continuous distribution by present in numpy

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Total population distribution

```python
# population 

normal = np.random.normal(0, 1, 1000)

geometric = np.random.geometric(0.2,1000)

hypergeometric = np.random.hypergeometric(500, 500, 20, 1000)

poisson = np.random.poisson(5, 1000)

binomial = np.random.binomial(10, 0.5, 1000)

exponential = np.random.exponential(1, 1000)

chi_squared = np.random.chisquare(2, 1000)

uniform = np.random.uniform(-1, 1, 1000)

lognormal =np.random.lognormal(1,0.5,1000)

# Plot data
fig, axs = plt.subplots(3, 3, figsize=(15, 10))

# normal
sns.histplot(normal, kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Normal')
print('mean_normal = ',normal.mean(), '\nstd_normal  = ',normal.std())

# geometric
sns.histplot(geometric, kde=True, ax=axs[0, 1])
axs[0, 1].set_title('geometric')
print('mean_geometric = ',geometric.mean(), '\nstd_geometric  = ',geometric.std())


#hypergeometric
sns.histplot(hypergeometric, kde=True, ax=axs[0, 2])
axs[0, 2].set_title('hypergeometric')
print('mean_hypergeometric = ',hypergeometric.mean(), '\nsdt_hypergeometric  = ',hypergeometric.std())

# poisson
sns.histplot(poisson, kde=True, ax=axs[1, 0])
axs[1, 0].set_title('Poisson')
print('mean_poisson = ',poisson.mean(), '\nstd_poisson  = ',poisson.std())

# binomial
sns.histplot(binomial, kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Binomial')
print('mean_binomial = ',binomial.mean(), '\nstd_binomial  = ',binomial.std())

# exponential
sns.histplot(exponential, kde=True, ax=axs[1, 2])
axs[1, 2].set_title('Exponential')
print('mean_exponential = ',exponential.mean(), '\nstd_exponential  = ',exponential.std())

# chi_squared
sns.histplot(chi_squared, kde=True, ax=axs[2, 0])
axs[2, 0].set_title('Chi-squared')
print('mean_chi_squared = ',chi_squared.mean(), '\nstd_chi_squared  = ',chi_squared.std())

# uniform
sns.histplot(uniform, kde=True, ax=axs[2, 1])
axs[2, 1].set_title('Uniform')
print('mean_uniform = ',uniform.mean(), '\nstd_uniform  = ',uniform.std())


# lognormal
sns.histplot(lognormal, kde=True, ax=axs[2, 2])
axs[2, 2].set_title('lognormal')
print('mean_lognormal = ',lognormal.mean(), '\nstd_lognormal  = ',lognormal.std())


plt.subplots_adjust(wspace=0.4)
plt.show()
```

# Sampling dsitribution

```python
# sample

num_sample = 499
print(f'No.of samples      = {num_sample}')
sample_size = 50
print(f'No.of samples_size = {sample_size}')

normal = np.random.normal(0, 1, 1000)
normal_sample = []
for x in range(num_sample):
        sample=np.random.choice(normal,sample_size)
        normal_sample.append(np.mean(sample))
        

geometric = np.random.geometric(0.2,1000)
geometric_sample = []
for x in range(num_sample):
        sample=np.random.choice(geometric,sample_size)
        geometric_sample.append(np.mean(sample))

hypergeometric = np.random.hypergeometric(500, 500, 20, 1000)
hypergeometric_sample = []
for x in range(num_sample):
        sample=np.random.choice(hypergeometric,sample_size)
        hypergeometric_sample.append(np.mean(sample))

poisson = np.random.poisson(5, 1000)
poisson_sample = []
for x in range(num_sample):
        sample=np.random.choice(poisson,sample_size)
        poisson_sample.append(np.mean(sample))

binomial = np.random.binomial(10, 0.5, 1000)
binomial_sample = []
for x in range(num_sample):
        sample=np.random.choice(binomial,sample_size)
        binomial_sample.append(np.mean(sample))

exponential = np.random.exponential(1, 1000)
exponential_sample = []
for x in range(num_sample):
        sample=np.random.choice(exponential,sample_size)
        exponential_sample.append(np.mean(sample))

chi_squared = np.random.chisquare(2, 1000)
chi_squared_sample = []
for x in range(num_sample):
        sample=np.random.choice(chi_squared,sample_size)
        chi_squared_sample.append(np.mean(sample))

uniform = np.random.uniform(-1, 1, 1000)
uniform_sample = []
for x in range(num_sample):
        sample=np.random.choice(uniform,sample_size)
        uniform_sample.append(np.mean(sample))

lognormal = np.random.lognormal(1,0.5,1000)
lognormal_sample = []
for x in range(num_sample):
        sample=np.random.choice(lognormal,sample_size)
        lognormal_sample.append(np.mean(sample))
        
        


# Plot data
fig, axs = plt.subplots(3, 3, figsize=(15, 10))

# normal
sns.histplot(normal_sample, kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Normal_sample')
print('mean_normal_sample = ',np.mean(normal_sample), '\nstd_normal_sample  = ',np.std(normal_sample))

# geometric
sns.histplot(geometric_sample, kde=True, ax=axs[0, 1])
axs[0, 1].set_title('geometric_sample')
print('mean_geometric_sample = ',np.mean(geometric_sample), '\nstd_geometric_sample  = ',np.std(geometric_sample))


#hypergeometric
sns.histplot(hypergeometric_sample, kde=True, ax=axs[0, 2])
axs[0, 2].set_title('hypergeometric_sample')
print('mean_hypergeometric_sample = ',np.mean(hypergeometric_sample), '\nsdt_hypergeometric_sample  = ',np.std(hypergeometric_sample))

# poisson
sns.histplot(poisson_sample, kde=True, ax=axs[1, 0])
axs[1, 0].set_title('poisson_sample')
print('mean_poisson_sample = ',np.mean(poisson_sample), '\nstd_poisson_sample  = ',np.std(poisson_sample))

# binomial
sns.histplot(binomial_sample, kde=True, ax=axs[1, 1])
axs[1, 1].set_title('binomial_sample')
print('mean_binomial_sample = ',np.mean(binomial_sample), '\nstd_binomial_sample  = ',np.std(binomial_sample))

# exponential
sns.histplot(exponential_sample, kde=True, ax=axs[1, 2])
axs[1, 2].set_title('exponential_sample')
print('mean_exponential_sample = ',np.mean(exponential_sample), '\nstd_exponential_sample  = ',np.std(exponential_sample))

# chi_squared
sns.histplot(chi_squared_sample, kde=True, ax=axs[2, 0])
axs[2, 0].set_title('chi_squared_sample')
print('mean_chi_squared_sample = ',np.mean(chi_squared_sample), '\nstd_chi_squared_sample  = ',np.std(chi_squared_sample))

# uniform
sns.histplot(uniform_sample, kde=True, ax=axs[2, 1])
axs[2, 1].set_title('uniform_sample')
print('mean_uniform_sample = ',np.mean(uniform_sample), '\nstd_uniform_sample  = ',np.std(uniform_sample))


# lognormal
sns.histplot(lognormal_sample, kde=True, ax=axs[2, 2])
axs[2, 2].set_title('lognormal_sample')
print('mean_lognormal_sample = ',np.mean(lognormal_sample), '\nstd_lognormal_sample  = ',np.std(lognormal_sample))


plt.subplots_adjust(wspace=0.4)
plt.show()
```

# Randomly selected distribution from given list
### list_dist = ['normal','geometric','hypergeometric','poisson','binomial','exponential','chi_squared','uniform','lognormal']

```python
def distribution():
    
    list_dist = ['normal','geometric','hypergeometric','poisson','binomial','exponential','chi_squared','uniform','lognormal']
    x = np.random.choice(list_dist)
    

    
    if x ==  'normal':
        y = np.random.normal(0, 1, 1000)
    elif x == 'geometric':
        y = np.random.geometric(0.2,1000)
    elif x == 'hypergeometric':
        y = np.random.hypergeometric(500, 500, 20, 1000)
    elif x == 'poisson':
        y = np.random.poisson(5, 1000)
    elif x == 'binomial':
        y = np.random.binomial(10, 0.5, 1000)
    elif x == 'exponential':
        y = np.random.exponential(1, 1000)
    elif x == 'chi_squared':
        y = np.random.chisquare(2, 1000)
    elif x == 'uniform':
        y = np.random.uniform(-1, 1, 1000)  
    else:
        y = np.random.lognormal(1,0.5,1000)
        
        
    
    print(f'p_mean     : {np.mean(y)}')
    print(f'p_variance : {np.var(y)}')
    print(f'p_std      : {np.std(y)}')
        
    sns.histplot(y, kde=True)
    plt.title(f'{x} Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
    
    
    num_sample = 499
    print(f'No.of samples : {num_sample}')
    
    sample_size = 50
    print(f'samples size  : {sample_size}')
    
    sample = []
    for i in range(num_sample):
        z = np.random.choice(y,sample_size)
        sample.append(np.mean(z))
    
    sns.histplot(sample, kde=True)
    plt.title(f'{x} sample_distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
    
    
    print(f's_mean        : {np.mean(sample)}')
    print(f's_variance    : {np.var(sample)}')
    print(f's_std         : {np.std(sample)}')
    print(f'satnded_error : {np.std(sample) / np.sqrt(sample_size)}')
    
    
distribution()
```
